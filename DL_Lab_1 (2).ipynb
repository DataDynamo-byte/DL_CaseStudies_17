```python
import tensorflow as tf
import numpy as np
print("tensor version: ", tf.__version__)
```

```python
# Scalar tensor
scalar = tf.constant(10) # --- scaler value ---
print(scalar)
```

```python
# 1D tensor -- Vector tensor
vector = tf.constant([1, 2, 3, 4]) # --- vector values ---
print(vector)
```

```python
# 2D Tensor -- Matrix tensor
matrix = tf.constant([[1, 2], [3, 4]]) # --- matrix values ---
print(matrix)
```

```python
#Operations on Scalar Tensor

a = tf.constant(10)
b = tf.constant(5)
print("Addition:", tf.add(a, b))
print("Subtraction:", tf.subtract(a, b))
print("Multiplication:", tf.multiply(a, b))
print("Division:", tf.divide(a, b))
```

```python
# Operations on  Vector Tensor

a = tf.constant([10, 20, 30])
b = tf.constant([1, 2, 3])

print("Addition:", tf.add(a, b))
print("Subtraction:", tf.subtract(a, b))
print("Multiplication:", tf.multiply(a, b))
print("Division:", tf.divide(a, b))
```

```python
# Operations on Matrix Tensor
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])

print("Addition:", tf.add(a, b))
print("Subtraction:", tf.subtract(a, b))
print("Multiplication:", tf.multiply(a, b))
print("Division:", tf.divide(a, b))
```

```python
#Data Types and Shape of Tensors

tensor = tf.constant([[1, 2, 3], [4, 5, 6]])

print("Data Type:", tensor.dtype)
print("Shape:", tensor.shape)
print("Number of Dimensions:", tf.rank(tensor)) # rank of matrix
print("Number of Dimensions:", tensor.ndim) # dimensions of matrix
```

```python
#Converting NumPy Arrays to Tensors

numpy_array = np.array([[1, 2, 3], [4, 5, 6]])

tensor = tf.convert_to_tensor(numpy_array)

print("NumPy Array:\n", numpy_array) #numpy array
print("TensorFlow Tensor:\n", tensor) #tensorflow tensor
```
